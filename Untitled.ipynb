```python
import pandas as pd
import matplotlib.pyplot as plt

species = ["V_cholerae","V_alginolyticus","V_parahaemolyticus","V_vulnificus"]
counts = [153,88,81,32]

plt.figure(figsize=(6,4))
plt.bar(species, counts)
plt.xlabel("Especie")
plt.ylabel("NÃºmero de genes")
plt.title("Genes de resistencia detectados por especie (AMRFinderPlus)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("analysis/resistome_genes_totales.png", dpi=300)
plt.show()
```

```python
unique_counts = [28,16,13,9]

plt.figure(figsize=(6,4))
plt.bar(species, unique_counts)
plt.xlabel("Especie")
plt.ylabel("Genes Ãºnicos")
plt.title("Diversidad genÃ©tica del resistoma por especie")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("analysis/resistome_genes_unicos.png", dpi=300)
plt.show()
```
