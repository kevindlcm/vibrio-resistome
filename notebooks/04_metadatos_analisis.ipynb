```python
import json
import glob
import pandas as pd
import os

META_DIR = "../metadata/"
OUT = "../analysis/"

rows = []

for f in glob.glob(META_DIR + "V_*/*.jsonl"):
    especie = os.path.basename(os.path.dirname(f))

    with open(f) as fh:
        for line in fh:
            j = json.loads(line)

            rows.append({
                "Species": especie,
                "Accession": j.get("accession"),
                "Country": j.get("country"),
                "Collection_year": j.get("collection_date"),
                "Isolation_source": j.get("isolation_source"),
                "Host": j.get("host")
            })

meta = pd.DataFrame(rows)

meta.to_csv(OUT + "metadata_vibrio.tsv", sep="\t", index=False)

meta.head()
```

```python
amr = pd.read_csv("../analysis/amrfinder_clean.tsv", sep="\t")
meta = pd.read_csv("../analysis/metadata_vibrio.tsv", sep="\t")

df = amr.merge(meta, on="Species", how="left")

df.head()
```

```python
import matplotlib.pyplot as plt

tabla = (
    df.groupby(["Isolation_source","Species"])
      .size()
      .unstack(fill_value=0)
)

tabla.plot(kind="bar", stacked=True, figsize=(8,5))

plt.ylabel("Número de genes RAM")
plt.tight_layout()
plt.savefig("../analysis/ram_por_fuente.png", dpi=300)
plt.show()
```
