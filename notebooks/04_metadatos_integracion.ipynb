```python
import json
import glob
import pandas as pd
import os
```

```python
META_DIR = "../metadata/"
OUT = "../analysis/"

rows = []

for f in glob.glob(META_DIR + "V_*/*.jsonl"):
    especie = os.path.basename(os.path.dirname(f))

    with open(f) as fh:
        for line in fh:
            j = json.loads(line)

            rows.append({
                "Species": especie,
                "Accession": j.get("accession"),
                "Country": j.get("country"),
                "Collection_year": j.get("collection_date"),
                "Isolation_source": j.get("isolation_source"),
                "Host": j.get("host")
            })

meta = pd.DataFrame(rows)

meta.to_csv(OUT + "metadata_vibrio.tsv", sep="\t", index=False)

meta.head()
```

```python
amr = pd.read_csv("../analysis/amr_master.tsv", sep="\t")

df = amr.merge(meta, on="Species", how="left")

df.to_csv("../analysis/amr_with_metadata.tsv", sep="\t", index=False)

df.head()
```
