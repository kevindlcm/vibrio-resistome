```python
import pandas as pd
import glob
import os
```

```python
BASE = "../analysis/resistome/"
OUT  = "../analysis/"

import os
os.makedirs(OUT, exist_ok=True)
```

```python
archivos = glob.glob(BASE + "V_*_amrfinder.tsv")

dfs = []

for f in archivos:
    df = pd.read_csv(f, sep="\t")
    especie = os.path.basename(f).replace("_amrfinder.tsv", "")
    df["Species"] = especie
    dfs.append(df)

amr = pd.concat(dfs, ignore_index=True)

amr.head()
```

```python
amr.to_csv(OUT + "amr_master.tsv", sep="\t", index=False)
```

```python
tabla = pd.crosstab(
    amr["Species"],
    amr["Element symbol"]
)

matriz = (tabla > 0).astype(int)

matriz.head()
```

```python
matriz.to_csv(OUT + "matriz_presencia_ausencia_amr.csv")
```
